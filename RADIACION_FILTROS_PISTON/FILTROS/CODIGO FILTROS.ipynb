FILTRO RLC PASA BAJOS
==============



***MAGNITUD Y FASE***


```python
import numpy as np
f = 10000
L= 2.25*10**-3
C= 112.54*10**-9
R= 200

w = 2 * np.pi * f  

s = 1j * w  

numerador = 1 / (L * C)
denominador = s**2 + (R/L) * s + 1/(L*C)

H_s = numerador / denominador

magnitud = np.abs(H_s)
magnitud_dB = 20 * np.log10(magnitud)
fase = np.angle(H_s)

print("la magnitud en la frecuencia de : ", f, "Hz es: ", magnitud)
print("la magnitud en dB la frecuencia de : ", f, "Hz es: ", magnitud_dB)
print("la fase en radianes en la frecuencia de : ", f, "Hz es: ", fase)
```

***PLOT DE MAGNITUD Y FASE***


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

f = np.linspace(20, 20000, 2000)
L= 2.25*10**-3
C= 112.54*10**-9
R= 200

w = 2 * np.pi * f  

s = 1j * w  

numerador = 1 / (L * C)
denominador = s**2 + (R/L) * s + 1/(L*C)

H_s = numerador / denominador

magnitud = np.abs(H_s)
fase = np.angle(H_s)



# Graficar la magnitud
plt.figure()
plt.semilogx(f, 20 * np.log10(magnitud))
plt.title('Magnitud de la Funcin de Transferencia')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud (dB)')
plt.grid(True)
plt.xticks([20,40,100, 200, 500, 1000, 2000, 4000,10000, 20000])
plt.gca().xaxis.set_major_formatter(ScalarFormatter())

# Graficar la fase
plt.figure()
plt.semilogx(f, np.degrees(fase))
plt.title('Fase de la Funcin de Transferencia')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Fase (grados)')
plt.grid(True)
plt.xticks([20,40,100, 200, 500, 1000, 2000, 4000,10000, 20000])


plt.gca().xaxis.set_major_formatter(ScalarFormatter())
plt.show()
```
